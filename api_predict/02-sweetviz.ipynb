```python
import pandas as pd

df = pd.read_csv('../data_set/SBAnational-EDA.csv')
df = df.drop(columns=['Unnamed: 0'])
display(df.head(1))
```

```python
from sklearn.preprocessing import LabelEncoder

# Créer un objet LabelEncoder
label_encoder = LabelEncoder()

# Itérer sur les colonnes de votre DataFrame
for colonne in df.columns:
    # Vérifier si la colonne est de type 'object' (catégorielle)
    if df[colonne].dtype == 'object':
        # Appliquer LabelEncoder à la colonne catégorielle
        df[colonne] = label_encoder.fit_transform(df[colonne])

# Afficher le DataFrame avec les colonnes encodées
display(df.head())
```

```python
import sweetviz as sv

my_report = sv.analyze(df, target_feat='MIS_Status')
my_report.show_html() # Default arguments will generate to "SWEETVIZ_REPORT.html"
```
